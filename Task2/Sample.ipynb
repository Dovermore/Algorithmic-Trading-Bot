```python
import numpy as np
```

```python
exp_return = np.array([[10, 0, 0, 5],[0,2.5,7.5,5],[7.5,7.5,2.5,5],[2.5,10,10,5]]) # Return table
p1 = np.linspace(0, 1, 101)  # Asset 1
p2 = np.array([np.arange(0, 1-p, 0.01) for p in p1])  # Asset 2
p3 = 1 - p2 - p1 # Asset 3
p4 = 0 # Risk free
```

```python
cov_matrix = np.cov(exp_return.T)

def performance(ps, b=0.01):
    ret = np.dot(exp_return, ps)
    variance = np.var(ret)
    return [np.mean(ret) - variance * b, np.mean(ret), variance, ps]
```

```python
all_ps = []

for i in range(p1.shape[0]):
    for j in range(p2[i].shape[0]):
        ps = [p1[i], p2[i][j], p3[i][j], 0]
        all_ps.append(ps)
all_ps = np.array(all_ps)
print(all_ps)
```

```python
sorted([performance(ps) for ps in all_ps], key=lambda x: x[0], reverse=True)
```
